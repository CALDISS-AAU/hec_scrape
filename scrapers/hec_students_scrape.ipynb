```python
url = 'https://www.hec.gov.pk/english/services/students/PCD/Pages/PCD.aspx' #main URL
html = requests.get(url).content #extract HTML
sel = Selector(text = html) #create Selector object
```

```python
table_sel = sel.css('div#MSOZoneCell_WebPartWPQ2 ::text').extract()
```

```python
## Scrape second page
driver = webdriver.Chrome(executable_path = 'C:\\bin\\chromedriver')

driver.get(url)
time.sleep(5)

driver.find_element_by_css_selector("#bottomPagingCellWPQ2 > table:nth-child(1) > tbody:nth-child(1) > tr:nth-child(1) > td:nth-child(2) > a:nth-child(1) > img:nth-child(1)").click()
time.sleep(3)

pageSource = driver.page_source

## Text extraction should occur with each page! (inside while loop)

doc = lh.fromstring(pageSource)

tr_elements = doc.xpath("//tr[@class='ms-itmhover']/td[@class='ms-vb2']")
tr_altelements = doc.xpath("//tr[@class='ms-alternating ms-itmhover']/td[@class='ms-vb2']")

print(len(list(t.text_content() for t in tr_elements)),
     len(list(t.text_content() for t in tr_altelements)))

table_rows.append(tr_elements)
table_rows.append(tr_altelements)

## Loop for additional pages
#pageSource = {}
#pageno = 1
#pagestr = "page_" + str(pageno)
#pageSources[pagestr] = driver.page_source

while True:
    try:
        driver.find_element_by_css_selector("#bottomPagingCellWPQ2 > table:nth-child(1) > tbody:nth-child(1) > tr:nth-child(1) > td:nth-child(3) > a:nth-child(1) > img:nth-child(1)").click()
        time.sleep(3)
        pageSource = driver.page_source 

        doc = lh.fromstring(pageSource)

        tr_elements = doc.xpath("//tr[@class='ms-itmhover']/td[@class='ms-vb2']")
        tr_altelements = doc.xpath("//tr[@class='ms-alternating ms-itmhover']/td[@class='ms-vb2']")

        print(len(list(t.text_content() for t in tr_elements)),
             len(list(t.text_content() for t in tr_altelements)))

        table_rows.append(tr_elements)
        table_rows.append(tr_altelements)
    except (NoSuchElementException, WebDriverException, ElementNotInteractableException) as e:
        print("No 'Next' button found")
        break
driver.quit()
```
